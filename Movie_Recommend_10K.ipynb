```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/drive/MyDrive/ML/Recommender_system/Copy of top10K-TMDB-movies.csv')
```

```python
movies=movies[['id','title','overview','genre']]
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genre']
```

```python
movies.head()
```

```python
new_data=movies[['id','title','tags']]
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
movies.dropna(inplace=True)
```

```python
new_data.dropna(inplace=True)
```

```python
new_data.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=9985,stop_words='english')
```

```python
vectors=cv.fit_transform(new_data['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
  index=new_data[new_data['title']==movie].index[0]
  movie_list=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])[:6]
  for i in movie_list:
    print(new_data.iloc[i[0]].title)
```

```python
recommend('Parasite')
```

```python
recommend('GoodFellas')
```
